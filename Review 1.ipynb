```python
def find_TP(y, y_hat):
   # counts the number of true positives (y = 1, y_hat = 1)
   return sum((y == 'Yes') & (y_hat == 'Yes'))
def find_FN(y, y_hat):
   # counts the number of false negatives (y = 1, y_hat = 0) Type-II error
   return sum((y == 'Yes') & (y_hat == 'No'))
def find_FP(y, y_hat):
   # counts the number of false positives (y = 0, y_hat = 1) Type-I error
   return sum((y == 'No') & (y_hat == 'Yes'))
def find_TN(y, y_hat):
   # counts the number of true negatives (y = 0, y_hat = 0)
   return sum((y == 'No') & (y_hat == 'No'))
```

```python
#'CM' Central Midfielder
fig, ax = plot_confusion_matrix(conf_mat=confusion_matrix1,class_names=class_names,figsize =(2, 2))
plt.title("Confusion Matrix of Central Midfielder\n")
plt.ylabel('Actual Labels\n')
plt.xlabel('\nPredicted Labels')

TP=(find_TP(y_test1,pre1))
TN=(find_TN(y_test1,pre1))
FN=(find_FN(y_test1,pre1))
FP=(find_FP(y_test1,pre1))
print('True Positive  :',TP)
print('True Negative  :',TN)
print('False Positive :',FP)
print('False Negative :',FN)
# Yes
accuracy=(TP+TN)/(TP+FP+TN+FN)
print('\nAccuracy       :',accuracy)
precision_yes = TP/(TP+FP)
print('Precision Yes  :',precision_yes)
recall_yes = TP/(TP+FN)
print('Recall Yes     :',recall_yes)
f1_score_yes = 2*((precision_yes*recall_yes)/(precision_yes+recall_yes))
# No
print('F1 score Yes   :',f1_score_yes)
precision_no = TN/(TN+FN)
print('Precision No   :',precision_no)
recall_no = TN/(TN+FP)
print('Recall No      :',recall_no)
f1_score_no = 2*((precision_no*recall_no)/(precision_no+recall_no))
print('F1 score No    :',f1_score_no)
```
